```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('punkt')
stop_words = set(stopwords.words("english"))
```

```python
import string
string.punctuation
```

```python
def rem_punc(text):
  new_text="".join([c for c in text if c not in string.punctuation])
  return new_text
df['clean_data']=df['Technology Skills'].apply(lambda x: rem_punc(x))
```

```python
df['lower_text']=df['clean_data'].apply(lambda x:x.lower())
```

```python
df['tokenize'] = df.apply(lambda row: nltk.word_tokenize(row['lower_text']),axis=1)
```

```python
stopwords=nltk.corpus.stopwords.words('english')
```

```python
def remove_stopwords(txt):
  text=[word for word in txt if word not in stopwords]
  return text
df['stop_free']=df['tokenize'].apply(lambda x: remove_stopwords(x))  
```

```python
df
```

```python
def rem_punc(text):
  new_text="".join([c for c in text if c not in string.punctuation])
  return new_text
af['clean_data']=af['RAW DATA'].apply(lambda x: rem_punc(x))
```

```python
af['lower_text1']=af['clean_data'].apply(lambda x:x.lower())
```

```python
af['tokenize1'] = af.apply(lambda row: nltk.word_tokenize(row['lower_text1']),axis=1)
```

```python
def remove_stopwords(txt):
  text=[word for word in txt if word not in stopwords]
  return text
af['stop_free1']=af['tokenize1'].apply(lambda x: remove_stopwords(x))
```

```python
af
```

```python
skills=['python']
for i in range(979):
  skills.extend(df['stop_free'][i])
```

```python
skills
```

```python
af['lable']=None
```

```python
for i in range(34116):
  for j in af['stop_free1'][i]:
    for k in skills:
      if j==k:
        af['lable'][i]="technical"
```
